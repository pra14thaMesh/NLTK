```python
import cv2
```

```python

# imread_color -- 1; imread_grayscale -- 0; imread_unchanged -- -1
# Load an color image in grayscale
img = cv2.imread(r'President_Barack_Obama.jpg',flags=0)
```

```python
#img = cv2.imread('many-faces-family.jpg')
# display an image
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#save an image
cv2.imwrite('Barack.jpg',img)
```

```python
#resizing the image using the scaling factor
img = cv2.imread('President_Barack_Obama.jpg')

res = cv2.resize(img,None,fx=2.0, fy=2.0, interpolation = cv2.INTER_CUBIC)
#res = cv2.resize(img,None,fx=0.4, fy=0.4, interpolation = cv2.INTER_AREA)
cv2.imshow('img',res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#resizing the image using the dimensions
dim = (500, 600)
print(img.shape)
print(dim)

resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
cv2.imshow("resized", resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Translation
import numpy as np

img = cv2.imread('President_Barack_Obama.jpg',0)
rows,cols = img.shape
#creating translation matrix
M = np.array([[1,0,-300],[0,1,-200]],dtype=np.float)
dst = cv2.warpAffine(img,M,(cols,rows))
print (M)
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()

#cv2.imwrite('Barack1.jpg',dst)
```

```python
# Rotation

img = cv2.imread('family.jpg')
rows,cols,channel = img.shape

M = cv2.getRotationMatrix2D((cols/2,rows/2),180,1)
dst = cv2.warpAffine(img,M,(cols,rows))

cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Blurring the image

img = cv2.imread('family.jpg')

blur1 = cv2.blur(img,(5,10))

cv2.imshow('img',blur1)
cv2.waitKey(0)
cv2.destroyAllWindows()

#cv2.imwrite('Barack1.jpg',blur1)
```

```python
#Canny Edge detection

import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('President_Barack_Obama.jpg',0)
edges = cv2.Canny(img,50,50)

plt.subplot(121)
plt.imshow(img,cmap='gray')
plt.title('Original Image')
plt.subplot(122)
plt.imshow(edges,cmap='gray')
plt.title('Edge Image')
plt.show()
```
